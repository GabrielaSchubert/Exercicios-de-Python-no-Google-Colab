1. Leia atentamente o que está descrito abaixo.

• O Ressaca’s é o bar de maior sucesso da região. A razão é a variada oferta de drinks a disposição dos
clientes.
• Por outro lado, o sistema de pedidos do bar se tornou mais complexo e está difícil mantê-lo por conta
da grande variedade de drinks.
• O problema é que a pedido dos clientes, novos drinks foram criados pela adição de diferentes aditivos
aos drinks existentes
• Os projetistas criaram classes para representar os novos drinks e calcular os preços de cada um deles
levando em conta os diferentes ingredientes
• Em decorrência houve uma verdadeira explosão de classes para representar os diferentes tipos de
drinks
Baseado na descrição acima, faça o que se pede:
• Apresentar o seu projeto (Diagrama de Classes) e implementar o sistema Drinks do Ressaca's Bar,
usando o padrão Decorator
• Imagine que você tem um drink chamado caipira, que custa 20,0 reais, porém quando o cliente vai
montando a sua caipira ele vai aumentando o preço
• Quero uma caipira de saquê, abacaxi, kiwi e açúcar
• Quero uma caipira de vodka, morango e adoçante
• Quero uma caipirinha (que é um drink pronto patenteado que não pode ser alterado – cachaça,
limão, gelo e açúcar)

Decorator.drawio.png

```python
from abc import ABC, abstractmethod

class Drink(ABC):
    @abstractmethod
    def get_description(self):
        pass

    @abstractmethod
    def cost(self):
        pass

class Caipira(Drink):
    def get_description(self):
        return "Caipira básica "

    def cost(self):
        return 20.0

class Caipirinha(Drink):
    def get_description(self):
        return "Caipirinha: (Cachaça, Limão, Gelo e Açúcar)"

    def cost(self):
        return 30.0

class IngredientDecorator(Drink):
    @abstractmethod
    def get_description(self):
        pass

class Saque(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Saquê"

    def cost(self):
        return self._drink.cost() + 5.0

class Vodka(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Vodka"

    def cost(self):
        return self._drink.cost() + 5.0

class Abacaxi(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Abacaxi"

    def cost(self):
        return self._drink.cost() + 2.0

class Kiwi(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Kiwi"

    def cost(self):
        return self._drink.cost() + 2.5

class Acucar(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Açúcar"

    def cost(self):
        return self._drink.cost() + 0.5

class Adocante(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Adoçante"

    def cost(self):
        return self._drink.cost() + 0.8


class Morango(IngredientDecorator):
    def __init__(self, drink):
        self._drink = drink

    def get_description(self):
        return f"{self._drink.get_description()}, Morango"

    def cost(self):
        return self._drink.cost() + 4.0

if __name__ == "__main__":

    caipira1 = Caipira()
    caipira1 = Saque(caipira1)
    caipira1 = Abacaxi(caipira1)
    caipira1 = Kiwi(caipira1)
    caipira1 = Acucar(caipira1)

    print(f"{caipira1.get_description()} custa R$ {caipira1.cost():.2f}")

    caipira2 = Caipira()
    caipira2 = Vodka(caipira2)
    caipira2 = Morango(caipira2)
    caipira2 = Adocante(caipira2)

    print(f"{caipira2.get_description()} custa R$ {caipira2.cost():.2f}")

    caipirinha = Caipirinha()
    print(f"{caipirinha.get_description()} custa R$ {caipirinha.cost():.2f}")
```
