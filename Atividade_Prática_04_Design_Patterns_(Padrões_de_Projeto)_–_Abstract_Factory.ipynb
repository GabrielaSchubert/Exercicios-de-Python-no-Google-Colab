1. Desenvolva o código para os diagramas de classes das figuras abaixo, usando o padrão de design
“Abstract Factory”.

```python
from abc import ABC, abstractmethod

class Furniture(ABC):
    @abstractmethod
    def show_style(self):
        pass

    @abstractmethod
    def furniture_function(self):
        pass

class ClassicChair(Furniture):
    def show_style(self):
        print("Cadeira Clássica")

    def furniture_function(self):
        print("Função: Sentar com estilo clássico")

class ClassicCabinet(Furniture):
    def show_style(self):
        print("Armário Clássico")

    def furniture_function(self):
        print("Função: Guardar itens em estilo clássico")

class ClassicDiningTable(Furniture):
    def show_style(self):
        print("Mesa de Jantar Clássica")

    def furniture_function(self):
        print("Função: Refeições em estilo clássico")

class ContemporaryChair(Furniture):
    def show_style(self):
        print("Cadeira Contemporânea")

    def furniture_function(self):
        print("Função: Sentar com estilo moderno")

class ContemporaryCabinet(Furniture):
    def show_style(self):
        print("Armário Contemporâneo")

    def furniture_function(self):
        print("Função: Guardar itens com estilo moderno")

class ContemporaryDiningTable(Furniture):
    def show_style(self):
        print("Mesa de Jantar Contemporânea")

    def furniture_function(self):
        print("Função: Refeições em estilo moderno")

class ScandinavianChair(Furniture):
    def show_style(self):
        print("Cadeira Escandinava")

    def furniture_function(self):
        print("Função: Sentar com simplicidade")

class ScandinavianCabinet(Furniture):
    def show_style(self):
        print("Armário Escandinavo")

    def furniture_function(self):
        print("Função: Guardar itens com simplicidade")

class ScandinavianDiningTable(Furniture):
    def show_style(self):
        print("Mesa de Jantar Escandinava")

    def furniture_function(self):
        print("Função: Refeições com simplicidade")

class FurnitureFactory(ABC):
    @abstractmethod
    def create_chair(self):
        pass

    @abstractmethod
    def create_cabinet(self):
        pass

    @abstractmethod
    def create_dining_table(self):
        pass

class ClassicFactory(FurnitureFactory):
    def create_chair(self):
        return ClassicChair()

    def create_cabinet(self):
        return ClassicCabinet()

    def create_dining_table(self):
        return ClassicDiningTable()

class ContemporaryFactory(FurnitureFactory):
    def create_chair(self):
        return ContemporaryChair()

    def create_cabinet(self):
        return ContemporaryCabinet()

    def create_dining_table(self):
        return ContemporaryDiningTable()

class ScandinavianFactory(FurnitureFactory):
    def create_chair(self):
        return ScandinavianChair()

    def create_cabinet(self):
        return ScandinavianCabinet()

    def create_dining_table(self):
        return ScandinavianDiningTable()

def client(factory: FurnitureFactory):
    chair = factory.create_chair()
    chair.show_style()
    chair.furniture_function()

    cabinet = factory.create_cabinet()
    cabinet.show_style()
    cabinet.furniture_function()

    dining_table = factory.create_dining_table()
    dining_table.show_style()
    dining_table.furniture_function()

print("Scandinavian Factory: ")
client(ScandinavianFactory())
print("\nContemporary Factory: ")
client(ContemporaryFactory())
print("\nClassic Factory: ")
client(ClassicFactory())
```

2. Desenvolva o código para o diagrama de classes na figura abaixo, usando o padrão de design “Abstract
Factory”.

```python
from abc import ABC, abstractmethod

class MotorVehicle(ABC):
    @abstractmethod
    def create(self):
        pass

class ElectricVehicle(MotorVehicle):
    def create(self):
        pass

class FutureVehicleMotorcycle(MotorVehicle):
    def create(self):
        return "Future Vehicle Motorcycle"

class FutureVehicleElectricCar(ElectricVehicle):
    def create(self):
        return "Future Vehicle EletricCar"

class NextGenMotorcycle(MotorVehicle):
    def create(self):
        return "Next Gen Motorcycle"

class NextGenCar(MotorVehicle):
    def create(self):
        return "Next Gen Car"

class Corporation(ABC):
    @abstractmethod
    def create_motor_vehicle(self):
        pass

    @abstractmethod
    def create_electric_vehicle(self):
        pass

class FutureVehicleCorporation(Corporation):
    def create_motor_vehicle(self):
        return FutureVehicleMotorcycle()

    def create_electric_vehicle(self):
        return FutureVehicleElectricCar()

class NextGenCorporation(Corporation):
    def create_motor_vehicle(self):
        return NextGenMotorcycle()

    def create_electric_vehicle(self):
        return NextGenCar()

def client(corporation: Corporation):
    motor_vehicle = corporation.create_motor_vehicle()
    print(motor_vehicle.create())

    electric_vehicle = corporation.create_electric_vehicle()
    print(electric_vehicle.create())

print(FutureVehicleCorporation().create_motor_vehicle().create())
print(FutureVehicleCorporation().create_electric_vehicle().create())
print(NextGenCorporation().create_motor_vehicle().create())
print(NextGenCorporation().create_electric_vehicle().create())
```

3. Desenvolva o código para o diagrama de classes na figura abaixo, usando o padrão de design “Abstract
Factory”.

```python
from abc import ABC, abstractmethod

class Weapon(ABC):
  @abstractmethod
  def usefun_function(self):
      pass

class Armor(ABC):
  @abstractmethod
  def usefun_function(self):
      pass

class Sword(Weapon):
  def usefun_function(self):
      print("Usando uma espada")

class Axe(Weapon):
  def usefun_function(self):
      print("Usando um machado")

class MageFireball(Weapon):
  def usefun_function(self):
      print("Usando uma bola de fogo")

class BodyArmor(Armor):
  def usefun_function(self):
      print("Vestindo armadura corporal")

class OrcArmor(Armor):
  def usefun_function(self):
      print("Vestindo armadura orc")

class CloakArmor(Armor):
  def usefun_function(self):
      print("Vestindo armadura mago")

class AbstractFactory(ABC):
  @abstractmethod
  def create_weapon(self):
      pass

  @abstractmethod
  def create_armor(self):
      pass

class WarriorFactory(AbstractFactory):
  def create_weapon(self):
      return Sword()

  def create_armor(self):
      return BodyArmor()

class OrcFactory(AbstractFactory):
  def create_weapon(self):
      return Axe()

  def create_armor(self):
      return OrcArmor()

class MageFactory(AbstractFactory):
  def create_weapon(self):
      return MageFireball()

  def create_armor(self):
      return CloakArmor()

def client(factory: AbstractFactory):
    weapon = factory.create_weapon()
    weapon.usefun_function()

    armor = factory.create_armor()
    armor.usefun_function()

print("Warrior Factory: ")
client(WarriorFactory())
print("\nOrc Factory: ")
client(OrcFactory())
print("\nMage Factory:")
client(MageFactory())
```
